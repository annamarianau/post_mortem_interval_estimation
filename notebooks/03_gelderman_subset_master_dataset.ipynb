```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('max_colwidth', 700)
```

Aim of this notebook is to extract and process the Gelderman cohort from the master dataset. 

```python
# import master dataset
master_df = pd.read_pickle('./../data/master_dataset.pkl')
print(master_df.shape)
display(master_df.head())
```

```python
# extract Gelderman cohort
subset_df = master_df[master_df.true_SOD_G.notnull()].copy()
subset_df.drop(['true_BP', 'pred_BP', 'pred_BP_conf', 'pred_SOD_G', 'pred_SOD_G_conf', 'true_SOD_M', 
                'BP_of_true_SOD_M'], axis=1, inplace=True)
print(subset_df.shape)
display(subset_df.head())
display(subset_df.info())
```

```python
subset_df.describe()
```

```python
subset_df.info()
```

```python
# convert time related features to datetime
subset_df['correct_img_date'] = pd.to_datetime(subset_df['correct_img_date'], format='%Y-%m-%d')
subset_df['date_placed_ARF'] = pd.to_datetime(subset_df['date_placed_ARF'], format='%Y-%m-%d')
display(subset_df.info())
display(subset_df.head())
```

```python
# extract donor and date from img column
subset_df['donor_date'] = subset_df.img.str.split('.', expand=True)[0]
display(subset_df.head(3))
```

```python
subset_df.columns
```

```python
grp_subset_df = subset_df.groupby(['new_id','old_id', 'correct_img_date', 'date_placed_ARF', 'PMI_days', 'sex',
                                  'age_at_death', 'est_weight_lb', 'est_stature_in', 
                                   'cadaver_stature_cm', 'cadaver_weight_lbs', 'donor_date'])\
                                    ['true_SOD_G','BP_of_true_SOD_G'].agg(list).reset_index()
display(grp_subset_df.head())
print(grp_subset_df.shape)
```

```python
# split into dataframes by anatomical regiions
head_df = subset_df[subset_df.BP_of_true_SOD_G == 'head'].copy()
torso_df = subset_df[subset_df.BP_of_true_SOD_G == 'torso'].copy()
limbs_df = subset_df[(subset_df.BP_of_true_SOD_G == 'arm') | (subset_df.BP_of_true_SOD_G == 'legs')].copy()
print(subset_df.shape, head_df.shape, torso_df.shape, limbs_df.shape)
```

```python
# keep only one image per donor_date
head_df.drop_duplicates(subset=['donor_date'], inplace=True)
torso_df.drop_duplicates(subset=['donor_date'], inplace=True)
limbs_df.drop_duplicates(subset=['donor_date'], inplace=True)
print(head_df.shape, torso_df.shape, limbs_df.shape)
```

```python
# align dataframes on donor_date column
head_torso_merge = pd.merge(head_df , torso_df[['donor_date', 'img','true_SOD_G', 'BP_of_true_SOD_G']], 
                            how='inner', on='donor_date')
print(head_torso_merge.shape)

head_torso_limbs_merge = pd.merge(head_torso_merge, limbs_df[['donor_date', 'img','true_SOD_G', 'BP_of_true_SOD_G']], 
                                  how='inner', on='donor_date')
print(head_torso_limbs_merge.shape)
head_torso_limbs_merge.head()
```

```python
# rename and drop columns 
head_torso_limbs_merge.rename(columns={"img_x": "img_head", "true_SOD_G_x": "true_SOD_G_head",
                                      "img_y": "img_torso", "true_SOD_G_y": "true_SOD_G_torso",
                                      "img": "img_limbs", "true_SOD_G": "true_SOD_G_limbs"}, inplace=True)
head_torso_limbs_merge.drop(['BP_of_true_SOD_G_x', 'BP_of_true_SOD_G_y', 'BP_of_true_SOD_G'], axis=1, inplace=True)
display(head_torso_limbs_merge.head())
display(head_torso_limbs_merge.info())
```

```python
# write to pickle file 
head_torso_limbs_merge.to_pickle('./../data/gelderman/gelderman_master_dataset.pkl')
```
